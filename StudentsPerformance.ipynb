```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\rmurz\\Downloads\\StudentsPerformance.csv")
```

```python
df
```

```python
df_nums = df.select_dtypes(exclude='object')
```

```python
df_objs = df.select_dtypes(include='object')
```

```python
df_objs = pd.get_dummies(df_objs,drop_first=True)
```

```python
final_df = pd.concat([df_nums,df_objs],axis=1)
```

```python
final_df
```

```python
final_df.corr()['math score'].sort_values()
```

```python
passmark=35
```

```python
final_df['Percentage'] = (final_df['math score']+final_df['reading score']+final_df['writing score'])/3
```

```python
final_df
```

```python
X = final_df.drop('Percentage',axis=1)
y = final_df['Percentage']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_lr_pred = lr.predict(X_test)
mean_squared_error(y_test, y_lr_pred)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge()
```

```python
param_grid = {'alpha': [0.01, 0.1, 1, 10, 100]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_models = GridSearchCV(estimator=model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5)
```

```python
grid_models.fit(X_train,y_train)
```

```python
y_pred = grid_models.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=0.1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)
print(lasso_model.coef_)
```

```python
feature_names = X.columns  
import pandas as pd
pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': lasso_model.coef_
}).sort_values(by='Coefficient', key=abs, ascending=False)
```
